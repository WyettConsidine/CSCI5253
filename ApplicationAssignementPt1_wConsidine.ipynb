```python
import seaborn as sns
import matplotlib.pyplot as plt

skillCompDF.plot(kind = 'barh', title="Comparison of Personal and Recommened Profficiencies Across Various Skills")
plt.gca().invert_yaxis()
#add axis titles
plt.legend(bbox_to_anchor=(1.4, 1.0), loc = 'upper right')
plt.xlabel('Pofficiency Level')
plt.ylabel('')
plt.grid(axis = 'x', linewidth = '.25')
```

```python
above = skillCompDF[skillCompDF['Personal Prof']>skillCompDF['Recommended Prof']].count()[0]
at = skillCompDF[skillCompDF['Personal Prof']==skillCompDF['Recommended Prof']].count()[0]
below = skillCompDF[skillCompDF['Personal Prof']<skillCompDF['Recommended Prof']].count()[0]

plt.pie([above, at, below], explode = [0.1,0,0])
plt.legend(['Personal Skill Exceding Recommened','Personal Skill Meeting Recommened','Personal Skill Below Recommened'],
           bbox_to_anchor=(1.5, 1.0), loc = 'upper right')
plt.title('Personal vs Recommended Skill Proficiencies')
plt.show() 
```
